**Filtering under two or more condition**

We are going to use AND and OR operator to filter with more than one condition. Let’s assume that we want to see the passengers whose *Fare is lesser than 100* and *who are female*. We are going to create 2 new masks to complete that.

**follow this (below) solution**

```python
# finding female passengers having ticket fare < $100

# creating two mask
df_fare_mask = df["Fare"] < 100
df_sex_mask = df["Sex"] == "female"

# we are fetching those records whose index are true
df[df_fare_mask & df_sex_mask].head(2)
```

```python
# do this 
# find passenger's whose fare is > $500 or older than 70 years.
# Note : or operator is indicated as pipe ->    |
df_fare_mask = df["Fare"] < 500
df_age_mask = df["Age"]>70

df[df_fare_mask|df_age_mask].head(4) 
```

**Dealing with missing values**

There are lots of ways to deal with missing values but most common are 
**to “drop the tuple”** 
or 
to **“fill it with median”.** 

We are going to ignore the “Cabin” column since %70 of that column is missing. And we are going to fill the missing Ages with median value of that column.

```python
# drop the “Cabin” column
dfwithoutCabin = df.drop(labels = ["Cabin"], axis=1)
dfwithoutCabin.head()
# If you want to drop more than one column, then add it in the square brackets. For example:
# df.drop(labels = ["Cabin", "Name"], axis=1).head()
```

```python
# drop those rows which have No Age information
# use dropna()
# follow this code
# making new data frame with dropped NA values 
new_df = dfwithoutCabin.dropna(axis = 0, how ='any') 
# how='any' would mean any value is NA then drop the row
# axis=0 means check row wise
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html
  
# comparing sizes of data frames 
print("Old data frame length:", len(dfwithoutCabin), "\nNew data frame length:",  
       len(new_df), "\nNumber of rows with at least 1 NA value: ", 
       (len(dfwithoutCabin)-len(new_df))) 
```

```python
# Filling missing values with .fillna()

# fill the missing values with the median of Age column
# df['Age'] = df['Age'].fillna((df['Age'].median()))
# df['Age'].head()

# or 
df['Age'].fillna(df['Age'].median(),inplace = True)
df['Age'].head(20)

#df['Age'].median()
```
